```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/sxfrul/suds-n-sparkle-data/main/surveyform.csv'
df = pd.read_csv(url ,skipinitialspace = True)
# Dataset is now stored in a Pandas Dataframe

# displaying the DataFrame
df

# creating a basic histogram
```

```python
df = df.dropna()
df
```

```python
df["How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)"] = df["How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)"].str.lower()

df
```

```python
df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'] = df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'].str.replace('rm', '')
df
```

```python
df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'] = df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'].str.strip()
df
```

```python
df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'] = pd.to_numeric(df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'], errors='coerce')
df
```

```python
df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'] = df['How much are you willing to pay for your laundry per wash? \n(Price range: RM4.00 to RM8.00)'].round(2)
df
```

```python
df.to_csv('cleaned_survey.csv')
```
